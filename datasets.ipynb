```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_points
import random
import os

# Ensure reproducibility
np.random.seed(42)
random.seed(42)
```

# Building a linear dataset

```python
# Initialize an empty list to store the generated points
rows = []

# Generate 50 points with specific labels
for _ in range(100):
    x = 6 * np.random.random() - 3
    y = 6 * np.random.random() - 3
    z = int(x + y > 0.5)
    rows.append([x, y, z])

# Generate 10 points with random labels
for _ in range(10):
    x = 6 * np.random.random() - 3
    y = 6 * np.random.random() - 3
    z = random.randint(0, 1)
    rows.append([x, y, z])

# Convert the list of points into a DataFrame
df_linear = pd.DataFrame(rows, columns=['x1', 'x2', 'y'])

# Ensure the directory exists
os.makedirs('./data', exist_ok=True)

# Save the DataFrame to a CSV file
df_linear.to_csv('./data/linear.csv', index=False)

# Function to plot the points
plot_points(df_linear[['x1', 'x2']], df_linear['y'])
```

# Building a circular dataset

```python
# Initialize an empty list to store the generated points
rows = []

# Generate 50 points with specific labels
for _ in range(100):
    x = 6 * np.random.random() - 3
    y = 6 * np.random.random() - 3
    z = int(x**2 + y**2 < 2.8)
    rows.append([x, y, z])

# Generate 10 points with random labels
for _ in range(10):
    x = 6 * np.random.random() - 3
    y = 6 * np.random.random() - 3
    z = random.randint(0,1)
    rows.append([x, y, z])


# Convert the list of points into a DataFrame
df_circle = pd.DataFrame(rows, columns=['x1', 'x2', 'y'])

# Ensure the directory exists
os.makedirs('./data', exist_ok=True)

# Save the DataFrame to a CSV file
df_circle.to_csv('./data/circle.csv', index=False)

# Function to plot the points
plot_points(df_circle[['x1', 'x2']], df_circle['y'])
```

# Building a two-circles dataset

```python
# Initialize an empty list to store the generated points
rows = []

# Generate 50 points with specific labels
for _ in range(200):
    x = 6 * np.random.random() - 3
    y = 6 * np.random.random() - 3
    z = int((x+1)**2 + y**2 < 2 or (x-1)**2 + y**2 < 2)
    rows.append([x, y, z])

# Generate 10 points with random labels
for _ in range(20):
    x = 6 * np.random.random() - 3
    y = 6 * np.random.random() - 3
    z = random.randint(0,1)
    rows.append([x, y, z])


# Convert the list of points into a DataFrame
df_two_circles = pd.DataFrame(rows, columns=['x1', 'x2', 'y'])

# Ensure the directory exists
os.makedirs('./data', exist_ok=True)

# Save the DataFrame to a CSV file
df_two_circles.to_csv('./data/two_circles.csv', index=False)

# Function to plot the points
plot_points(df_two_circles[['x1', 'x2']], df_two_circles['y'])
```
